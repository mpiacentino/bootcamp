```python
section_means = df.groupby('Embryo')['Cntl Pax7 Tube Cells', 'Expt Pax7 Tube Cells']
section_means.describe()
```

```python
section_means = df.groupby('Embryo')['Cntl Pax7 Tube Cells', 'Expt Pax7 Tube Cells'].mean().reset_index()
section_means
```

```python
section_means['Expt/Cntl'] = section_means['Expt Pax7 Tube Cells'] / section_means['Cntl Pax7 Tube Cells']
section_means
```

```python
alt.Chart(section_means,
    title='heyhey'
    ).mark_circle(
    ).encode(
        x='Expt/Cntl:Q',
    )
```
