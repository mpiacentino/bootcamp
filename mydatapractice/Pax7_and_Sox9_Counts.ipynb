```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('smpd3_quants.csv')
```

```python
df.head()
```

```python
# hides all values that are not at HH stage 9, but does not remove them
mask = df['HH Stage'] != 9
df[~mask].head()
```

```python
# retain only rows that are from stage 9 embryos
stage_9 = df[df['HH Stage'] == 9]
stage_9.head()
```

```python
section_means = (stage_9.groupby(['Embryo'])
                ['Cntl Pax7 Total Cells', 'Expt Pax7 Total Cells', 
                 'Cntl Sox9 Cells', 'Expt Sox9 Cells'])
section_means_ratio = section_means.mean().reset_index()
section_means_ratio
```

```python
section_means_ratio['Expt/Cntl Pax7'] = (section_means_ratio['Expt Pax7 Total Cells'] 
                                         / section_means_ratio['Cntl Pax7 Total Cells'])
section_means_ratio['Expt/Cntl Sox9'] = (section_means_ratio['Expt Sox9 Cells'] 
                                         / section_means_ratio['Cntl Sox9 Cells'])
section_means_ratio
```

```python
alt.Chart(section_means_ratio,
    title='Normalized Cell Counts'
    ).mark_circle(
        color='blue'
    ).encode(
        x='Expt/Cntl Pax7:Q',
    )

#problem - I can't figure out how to plot Sox9 results as a separate dataset on the same plot
```

```python
pax7 = alt.Chart(section_means_ratio,
        title='Norm. Pax7 Counts',
        height=50,
        width=250,
    ).mark_circle(
        size=50,
        opacity=0.9
    ).encode(     
        x=alt.X('Expt/Cntl Pax7:Q',
               title='Expt/Cntl')
    )

sox9 = alt.Chart(section_means_ratio,
    title='Norm. Sox9 Counts',
    height=50,
    width=250
    ).mark_circle(
        size=50,
        opacity=0.9,
        color='orange'
    ).encode(
        x=alt.X('Expt/Cntl Sox9:Q',
               title='Expt/Cntl')
    )

pax7 + sox9
```

```python
section_means_ratio.to_csv('section_means_ratio.csv')
```
