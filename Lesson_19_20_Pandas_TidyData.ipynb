split, apply, combine occurs over and over and over
you're taking a split from your dataframe (pull out entries you want), apply is computing what you want from that, and combine will then put them back together 

```python
#this will group dataframe by the column insomnia, and the number of groups 
#are defined by the number of values found in column 'insomnia'
grouped = df.groupby('insomnia')
    #groupby is our split step
```

```python
grouped.median()
    #median is our apply AND combine
```

```python
grouped['percent correct'].median()
```

```python
grouped.median()
```

```python
grouped.median().reset_index()
```

```python
#define our own function to deal with things not present in pandas
def coeff_of_var(data):
    return np.std(data) / np.mean(data)
```

```python
grouped = df.groupby('insomnia')
```

```python
grouped.median()
```

```python
grouped.agg(coeff_of_var).reset_index()
```

```python
#transformations
df['rank grouped by insomnia'] = (grouped['percent correct']
                                    .rank(method='first'))

df
```

```python
#transformations
df_sorted = df.sort_values(by=['insomnia', 'percent correct'])

df_sorted[['insomnia', 'percent correct', 'rank grouped by insomnia']]
```

```python
#transformations
df_sorted = df.sort_values(by=['insomnia', 'percent correct'], ascending = False)

df_sorted[['insomnia', 'percent correct', 'rank grouped by insomnia']]
```
