```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('data/gfmt_sleep.csv', na_values='*')
df['insomnia'] = df['sci'] <= 16
```

```python
df.head()
```

Altair is a very high level graphics package, and really it needs one thing to function: a pd dataframe

```python
alt.Chart(df,
        height=400,
        width=500
    ).mark_point(
        size=75,
        filled=True,
        opacity=0.9
    ).encode(
        x='confidence when correct:Q',
        y='confidence when incorrect:Q',
        color='insomnia:N',
        shape='gender:N',
        tooltip=['gender', 'insomnia', 'percent correct']
    )
```

```python
#generate a good tidy dataframe that we can now put into altair
#do the aggregates in pandas because explicit better than implicit
df_summary = (df.groupby('ID')['impact force (mN)']
                 .agg(['mean', 'sem'])
                 .reset_index())

df_summary
```

```python
df_summary['error low'] = df_summary['mean'] - 1.96*df_summary['sem']
df_summary['error high'] = df_summary['mean'] + 1.96*df_summary['sem']

df_summary
```

```python
bars = alt.Chart(df_summary
        ).mark_bar(
        ).encode(
            x=alt.X('mean:Q', title='impact force (mN)'),
            y=alt.Y('ID:N', title='Frog ID')
        )

#rule - collects two points as a line, so you'll need to encode x and x2, or y and y2

error_bars = alt.Chart(df_summary
        ).mark_rule(
        ).encode(
            x='error low:Q',
            x2='error high:Q',
            y=alt.Y('ID:N', title='Frog ID')
        )

bars
error_bars
```
